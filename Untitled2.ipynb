```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout, Bidirectional
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

print("="*50)
print("LOADING SPAM DATASET")
print("="*50)

df = pd.read_csv("spam.csv", encoding='latin-1')
df = df[['v1', 'v2']]
df.columns = ['label', 'message']
df['label'] = df['label'].map({'ham': 0, 'spam': 1})

print(f"Total messages: {len(df)}")
print(f"Spam: {df['label'].sum()} | Ham: {len(df)-df['label'].sum()}")

print("\n" + "="*50)
print("PREPROCESSING TEXT")
print("="*50)

tokenizer = Tokenizer(num_words=5000)
tokenizer.fit_on_texts(df['message'])
X = pad_sequences(tokenizer.texts_to_sequences(df['message']), maxlen=100)
y = df['label'].values

split = int(len(df) * 0.8)
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

print(f"Training samples: {len(X_train)}")
print(f"Testing samples: {len(X_test)}")

print("\n" + "="*50)
print("BUILDING LSTM NEURAL NETWORK")
print("="*50)

model = Sequential([
    Embedding(5000, 64, input_length=100),
    Bidirectional(LSTM(64, dropout=0.2)),
    Dense(32, activation='relu'),
    Dropout(0.3),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()

print("\n" + "="*50)
print("TRAINING MODEL")
print("="*50)

history = model.fit(X_train, y_train, epochs=5, batch_size=32, validation_split=0.2)

print("\n" + "="*50)
print("EVALUATING MODEL")
print("="*50)

loss, accuracy = model.evaluate(X_test, y_test)
print(f"\n✅ SPAM DETECTION ACCURACY: {accuracy:.4f}")

print("\n" + "="*50)
print("TESTING WITH CUSTOM MESSAGES")
print("="*50)

test_messages = [
    "Congratulations! You won $1000! Click here to claim",
    "Hey, want to grab lunch tomorrow?",
    "URGENT! Your account will be closed. Verify now!",
    "I'll be home by 6pm"
]

for msg in test_messages:
    seq = tokenizer.texts_to_sequences([msg])
    pad = pad_sequences(seq, maxlen=100)
    pred = model.predict(pad, verbose=0)[0][0]
    result = "SPAM" if pred > 0.5 else "HAM"
    print(f"\nMessage: {msg}")
    print(f"Result: {result} (confidence: {pred:.2%})")

print("\n🎉 SPAM DETECTION COMPLETE!")
```

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout, Bidirectional
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

print("="*50)
print("LOADING IMDB DATASET")
print("="*50)

df = pd.read_csv("IMDB_Dataset.csv")
df['label'] = df['sentiment'].map({'positive': 1, 'negative': 0})

print(f"Total reviews: {len(df)}")
print(f"Positive: {df['label'].sum()} | Negative: {len(df)-df['label'].sum()}")

# Use 10,000 samples for faster training
df_sample = df.sample(10000, random_state=42)
print(f"Using sample: {len(df_sample)} reviews")

print("\n" + "="*50)
print("PREPROCESSING TEXT")
print("="*50)

tokenizer = Tokenizer(num_words=10000)
tokenizer.fit_on_texts(df_sample['review'])
X = pad_sequences(tokenizer.texts_to_sequences(df_sample['review']), maxlen=200)
y = df_sample['label'].values

split = int(len(df_sample) * 0.8)
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

print(f"Training samples: {len(X_train)}")
print(f"Testing samples: {len(X_test)}")

print("\n" + "="*50)
print("BUILDING BiLSTM NEURAL NETWORK")
print("="*50)

model = Sequential([
    Embedding(10000, 128, input_length=200),
    Bidirectional(LSTM(64, dropout=0.2)),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()

print("\n" + "="*50)
print("TRAINING MODEL (This may take 5-10 minutes)")
print("="*50)

history = model.fit(X_train, y_train, epochs=3, batch_size=64, validation_split=0.2)

print("\n" + "="*50)
print("EVALUATING MODEL")
print("="*50)

loss, accuracy = model.evaluate(X_test, y_test)
print(f"\n✅ SENTIMENT ANALYSIS ACCURACY: {accuracy:.4f}")

print("\n" + "="*50)
print("TESTING WITH CUSTOM REVIEWS")
print("="*50)

test_reviews = [
    "This movie was amazing! Best film ever!",
    "Worst movie ever. Complete waste of time.",
    "The acting was superb, I loved it!",
    "Terrible plot, boring characters."
]

for review in test_reviews:
    seq = tokenizer.texts_to_sequences([review])
    pad = pad_sequences(seq, maxlen=200)
    pred = model.predict(pad, verbose=0)[0][0]
    result = "POSITIVE" if pred > 0.5 else "NEGATIVE"
    print(f"\nReview: {review}")
    print(f"Sentiment: {result} (confidence: {pred:.2%})")

print("\n🎉 SENTIMENT ANALYSIS COMPLETE!")
```
